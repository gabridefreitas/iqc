## Operações com números complexos

### Chave 1

```python
a = 2+3j
b = 4-1j

print(a * b)
```

### Chave 2

```python
a = 6 + 2j
b = 2 - 1j

print(np.conj(a / b), a/b, (10+10j)/5)
```

### Chave 3

```python
a = 5-7j

print(np.conj(a))
```

### Chave 4

```python
matriz = np.array([[2, 1+1j], [3, -1j]])
print(np.transpose(matriz))

# print(matriz.T)
```

### Chave 5

```python
matriz = np.array([[2+1j, (2/3)*1j], [5*4-2j, 10-4j], [np.pi+4j, -2-1j], [-3, 1+5*5j]])

print(np.conjugate(np.transpose(matriz)))
```

### Chave 6

```python
a = np.array([[2+1j, -3j], [1+2j, -2+1j]])
b = np.array([[-3-1j, 1+1j, 4], [2, 2-1j, 5]])

print(a @ b)
```

### Chave 7

```python
a = np.array([[1], [1j]])
b = np.array([[2], [-1j]])


# Produto interno
print(np.conjugate(a.T) @ b)
```

### Chave 8

```python
a = np.array([[2], [-1j]])
b = np.array([[5], [2j]])


# Produto externo
print(a @ np.conjugate(b.T))
```

### Chave 9

```python
z = 3-5j

# Modulo Z
print(np.abs(z))
```
