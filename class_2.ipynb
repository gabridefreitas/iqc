## Operações com números complexos

### Multiplicação de números complexos

```python
a = 2 + 2j
b = 3 + 5j

print(f"A multiplicação de {a} e {b} é {a * b}")
```

### Divisão de números complexos

```python
a = 2 + 1j
b = 3 + 2j

print(f"A divisão de {a} e {b} é {a / b}")
print((b.real * a.real + b.imag * a.imag) / (b.real**2 + b.imag**2))
```

### Vetor coluna (ket) e vetor linha (bra)

```python
ket0 = np.array([[1], [0]])
bra0 = np.array([[1, 0]])

print(f"Vetor coluna (ket) |0>:\n{ket0}")
print(f"Vetor linha (bra) <0|:\n{bra0}")
```
